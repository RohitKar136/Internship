# 1) Write a python program to display all the header tags from wikipedia.org.

```python
page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
Content= BeautifulSoup(page.content)
Content
```

All The Header Tags

```python
Headers=[]
for i in Content.find_all('span', class_="mw-headline"):
    Headers.append(i.text)
Headers
```

```python
print(len(Headers))
```

# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page2= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
page3= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
page2
page3
```

```python
IMDB= BeautifulSoup(page2.content)
IMDB2= BeautifulSoup(page3.content)
IMDB
IMDB2
```

Steps For Year of Release

```python
YearofRelease=[]
for i in IMDB.find_all('span',class_="lister-item-year text-muted unbold"):
    YearofRelease.append(i.text)
YearofRelease
YearofRelease2=[]
for i in IMDB2.find_all('span',class_="lister-item-year text-muted unbold"):
    YearofRelease2.append(i.text)
print(YearofRelease2)
```

```python
print(len(YearofRelease), len(YearofRelease2))
```

```python
Year_Of_Release= YearofRelease + YearofRelease2
print(Year_Of_Release)
```

```python
len(Year_Of_Release)
```

Steps For Rating

```python
Rating1=[]
for i in IMDB.find_all('div',class_="inline-block ratings-imdb-rating"):
    Rating1.append(i.text)
Rating1
Rating2=[]
for i in IMDB2.find_all('div', class_="inline-block ratings-imdb-rating"):
    Rating2.append(i.text)
print(Rating2)
```

```python
print(len(Rating1), len(Rating2))
```

```python
Rating= Rating1 + Rating2
print(Rating)
```

```python
len(Rating)
```

Steps for Name

```python
Name1=[]
for i in IMDB.find_all('h3',class_="lister-item-header"):
    Name1.append(i.text)
Name1
```

```python
Name2=[]
for i in IMDB2.find_all('h3',class_="lister-item-header"):
    Name2.append(i.text)
Name2
```

```python
print(len(Name1), len(Name2))
```

```python
Name = Name1 + Name2
print(Name)
```

```python
print(len(Name), len(Rating), len(Year_Of_Release))
```

```python
import pandas as qstn2
df=qstn2.DataFrame({'Name':Name,'Rating':Rating,'Year_Of_Release':Year_Of_Release})
df
```

# 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp 

```python
page5= requests.get('https://meesho.com/bags%02ladies/pl/p7vbp')
page5
```

```python
Meesho= BeautifulSoup(page5.content)
Meesho
```

Product Name

```python
Name=[]
for i in Meesho.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    Name.append(i.text)
Name
len(Name)
```

Product Price

```python
Price=[]
for i in Meesho.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    Price.append(i.text)
Price
len(Price)
```

Product Discount

```python
Discount=[]
for i in Meesho.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    Discount.append(i.text)
Discount
len(Discount)
```

```python
import pandas as qstn4
df=qstn4.DataFrame({'Product_Name':Name,'Product_Price':Price,'Product_Discount':Discount})
df
```
