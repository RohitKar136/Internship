# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page1 =requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page1
```

```python
ODIT=BeautifulSoup(page1.content)
ODIT
```

Team Names

```python
Team=[]
for i in ODIT.find_all('span',class_="u-hide-phablet"):
    Team.append(i.text)
print(Team)
```

```python
Teams=Team[:10]
```

```python
print(Teams)
```

Matches

```python
Match1= ODIT.find('td',class_="rankings-block__banner--matches")
Match1.text
```

```python
Match2=[]
for i in ODIT.find_all('td',class_="table-body__cell u-center-text"):
    Match2.append(i.text)
print(Match2)
```

```python
v=[]
for i in range(0, len(Match2), 2) :
    v.append(Match2[i])
print(v)
    
```

```python
v.insert(0,Match1.text)
print(v)
```

```python
Matches=v[:10]
print(Matches)
```

Points

```python
point1= ODIT.find('td',class_="rankings-block__banner--points")
point1.text
```

```python
p=[]
for i in range(1, len(Match2), 2) :
    p.append(Match2[i])
print(p)
    
```

```python
p.insert(0,point1.text)
print(p)
```

```python
Points=p[:10]
print(Points)
```

Ratings

```python
rating1=ODIT.find('td',class_="rankings-block__banner--rating u-text-right")
rating1.text
```

```python
rating2=[]
for i in ODIT.find_all('td',class_="table-body__cell u-text-right rating"):
    rating2.append(i.text)
print(rating2)
```

```python
rating2.insert(0,rating1.text)
print(rating2)
```

```python
Rating=rating2[:10]
print(Rating)
```

```python
import pandas as qstn5a
df=qstn5a.DataFrame({'Teams':Teams,'Matches':Matches,'Points':Points,'Rating':Rating})
df
```
