# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.


```python
page1=requests.get('https://coreyms.com/')
page1
```

```python
coreyms=BeautifulSoup(page1.content)
coreyms
```

Heading

```python
Heading=[]
for i in coreyms.find_all('a',class_="entry-title-link"):
    Heading.append(i.text)
print(Heading)
```

Content

```python
Content=[]
for i in coreyms.find_all('div',class_="entry-content"):
    Content.append(i.text)
print(Content)
```

Date

```python
Date=[]
for i in coreyms.find_all('time',class_="entry-time"):
    Date.append(i.text)
print(Date)
```

```python
link=[]
for i in coreyms.find_all("iframe", class_="youtube-player"):
    link.append(i['src'])
link
```

```python
len(Heading),len(Date),len(Content),len(link)
```

```python
import pandas as qstn7
df=qstn7.DataFrame({'Heading':Heading,'Date':Date,'Content':Content,'Video_URL':link})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :i) Restaurant name ,ii) Cuisine ,iii) Location ,iv) Ratings ,v) Image URL

```python
page3=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page3
```

```python
restro=BeautifulSoup(page3.content)
restro
```

Name of Restaurant

```python
name=[]
for i in restro.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
name
```

Cuisine

```python
cuis=[]
for i in restro.find_all('span',class_="double-line-ellipsis"):
    cuis.append(i.text)
cuis
```

Location

```python
Location=[]
for i in restro.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
Location
```

Ratings

```python
Rating=[]
for i in restro.find_all('div',class_="restnt-rating rating-4"):
    Rating.append(i.text)
Rating
```

Image URL

```python
Image=[]
for i in restro.find_all('img',class_="no-img"):
    Image.append(i['data-src'])
Image
```

```python
import pandas as qstn9
df=qstn8.DataFrame({'Name_of_Restaurant':name,'Cuisine':cuis,'Location':Location,'Ratings':Rating,'Image_URL':Image})
df
```
